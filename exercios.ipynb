## Exercício 8

```python
medias_20 = {
    'ano': 2020,
    'overall_media': df_20['overall'].mean(),
    'potential_media': df_20['potential'].mean(),
    'age_media': df_20['age'].mean(),
    'total_jogadores': len(df_20)
}

medias_21 = {
    'ano': 2021,
    'overall_media': df_21['overall'].mean(),
    'potential_media': df_21['potential'].mean(),
    'age_media': df_21['age'].mean(),
    'total_jogadores': len(df_21)
}

df_comparacao_anos = pd.DataFrame([medias_20, medias_21])

df_comparacao_anos['overall_media'] = df_comparacao_anos['overall_media'].round(2)
df_comparacao_anos['potential_media'] = df_comparacao_anos['potential_media'].round(2)
df_comparacao_anos['age_media'] = df_comparacao_anos['age_media'].round(2)

print("=" * 80)
print("COMPARAÇÃO DE MÉDIAS GLOBAIS ENTRE ANOS:")
print("=" * 80)
print(df_comparacao_anos.to_string(index=False))

print("\n" + "=" * 80)
print("ANÁLISE DE VARIAÇÃO:")
print("=" * 80)

variacao_overall = df_comparacao_anos.loc[1, 'overall_media'] - df_comparacao_anos.loc[0, 'overall_media']
variacao_potential = df_comparacao_anos.loc[1, 'potential_media'] - df_comparacao_anos.loc[0, 'potential_media']
variacao_age = df_comparacao_anos.loc[1, 'age_media'] - df_comparacao_anos.loc[0, 'age_media']

print(f"Variação Overall 2020→2021: {variacao_overall:+.2f}")
print(f"Variação Potential 2020→2021: {variacao_potential:+.2f}")
print(f"Variação Age 2020→2021: {variacao_age:+.2f}")
print("=" * 80)
```

```python
engine_ex10 = create_engine("sqlite:///:memory:")

df_20.to_sql("jogadores_20", engine_ex10, index=False, if_exists="replace")
df_21.to_sql("jogadores_21", engine_ex10, index=False, if_exists="replace")

print("=" * 80)
print("BANCO SQL COM MÚLTIPLAS TABELAS CRIADO!")
print("=" * 80)
print("Tabelas: jogadores_20, jogadores_21")

query_ex10 = """
    SELECT DISTINCT 
        j20.short_name, 
        j20.long_name, 
        j20.overall AS overall_20,
        j21.overall AS overall_21,
        j20.potential AS potential_20,
        j21.potential AS potential_21
    FROM jogadores_20 j20
    INNER JOIN jogadores_21 j21 ON j20.short_name = j21.short_name
    LIMIT 100
"""

df_comuns = pd.read_sql(query_ex10, engine_ex10)

print(f"\nJOGADORES PRESENTES EM AMBOS OS ANOS (primeiros 100):")
print("-" * 80)
print(df_comuns.to_string(index=False))

print("\n" + "=" * 80)
print(f"Total de registros retornados: {len(df_comuns)}")
print("=" * 80)
```

## Exercício 12

```python
ANOS = {
    15: 'players_15.csv',
    16: 'players_16.csv',
    17: 'players_17.csv',
    18: 'players_18.csv',
    19: 'players_19.csv',
    20: 'players_20.csv',
    21: 'players_21.csv'
}

dfs_anos = {}

print("=" * 80)
print("CARREGANDO ARQUIVOS FIFA 15-21...")
print("=" * 80)

for ano, arquivo in ANOS.items():
    try:
        if os.path.exists(arquivo):
            df = pd.read_csv(arquivo)
            df_anchor = df[df['short_name'].isin(JOGADORES_ANCORA)].copy()
            dfs_anos[ano] = df_anchor
            print(f"✓ {arquivo}: {len(df_anchor)} jogadores âncora encontrados")
        else:
            print(f"✗ {arquivo}: Arquivo não encontrado")
    except Exception as e:
        print(f"✗ {arquivo}: Erro ao carregar - {str(e)}")

print("\n" + "=" * 80)
print("CONSOLIDANDO DADOS ENTRE ANOS...")
print("=" * 80)

primeiro_ano = min(dfs_anos.keys())
df_consolidado_ex12 = dfs_anos[primeiro_ano][['short_name', 'overall', 'potential', 'player_positions', 'age']].copy()
df_consolidado_ex12.rename(
    columns={
        'overall': f'overall_{primeiro_ano}',
        'potential': f'potential_{primeiro_ano}',
        'player_positions': 'posicao_principal'
    },
    inplace=True
)

for ano in sorted(dfs_anos.keys())[1:]:
    df_ano = dfs_anos[ano][['short_name', 'overall', 'potential']].copy()
    df_ano.rename(
        columns={
            'overall': f'overall_{ano}',
            'potential': f'potential_{ano}'
        },
        inplace=True
    )
    df_consolidado_ex12 = pd.merge(
        df_consolidado_ex12,
        df_ano,
        on='short_name',
        how='inner'
    )

if 21 in dfs_anos:
    df_idade = dfs_anos[21][['short_name', 'age']].copy()
    df_idade.rename(columns={'age': 'age_21'}, inplace=True)
    df_consolidado_ex12 = pd.merge(
        df_consolidado_ex12,
        df_idade,
        on='short_name',
        how='left'
    )

print(f"✓ DataFrame consolidado: {len(df_consolidado_ex12)} jogadores")

print("\nCALCULANDO EVOLUÇÃO...")

colunas_overall = [col for col in df_consolidado_ex12.columns if col.startswith('overall_')]
if len(colunas_overall) >= 2:
    primeira_col = sorted(colunas_overall)[0]
    ultima_col = sorted(colunas_overall)[-1]
    
    df_consolidado_ex12[f'evolucao_overall_{primeira_col.split("_")[1]}_21'] = (
        df_consolidado_ex12[ultima_col] - df_consolidado_ex12[primeira_col]
    )
    
    colunas_potential = [col for col in df_consolidado_ex12.columns if col.startswith('potential_')]
    primeira_pot = sorted(colunas_potential)[0]
    ultima_pot = sorted(colunas_potential)[-1]
    
    df_consolidado_ex12[f'evolucao_potential_{primeira_pot.split("_")[1]}_21'] = (
        df_consolidado_ex12[ultima_pot] - df_consolidado_ex12[primeira_pot]
    )

evolucao_col = [col for col in df_consolidado_ex12.columns if col.startswith('evolucao_overall_')][0]

def categorizar_tendencia(valor):
    if valor > 2:
        return 'subiu'
    elif valor < -2:
        return 'caiu'
    else:
        return 'estavel'

df_consolidado_ex12['tendencia_overall'] = df_consolidado_ex12[evolucao_col].apply(categorizar_tendencia)

print("✓ Evolução e tendência calculadas")

output_filename = 'relatorio_fifa_15_21_sj90.csv'
df_consolidado_ex12.to_csv(output_filename, index=False)
print(f"✓ Arquivo salvo: {output_filename}")
```

```python

print("\n" + "=" * 80)
print("RELATÓRIO FINAL - EVOLUÇÃO LONGITUDINAL FIFA 15-21:")
print("=" * 80)

colunas_display = ['short_name']
colunas_display.extend([col for col in df_consolidado_ex12.columns if col.startswith('overall_')])
colunas_display.extend([col for col in df_consolidado_ex12.columns if col.startswith('potential_')])
colunas_display.extend([col for col in df_consolidado_ex12.columns if col.startswith('evolucao_')])
colunas_display.append('tendencia_overall')

if 'age_21' in df_consolidado_ex12.columns:
    colunas_display.append('age_21')

df_display = df_consolidado_ex12[colunas_display].copy()

for col in df_display.columns:
    if df_display[col].dtype in ['float64', 'int64'] and col not in ['age_21']:
        df_display[col] = df_display[col].round(2)

print(df_display.to_string(index=False))

print("\n" + "=" * 80)
print(f"RESUMO ESTATÍSTICO:")
print("=" * 80)
print(f"Total de jogadores: {len(df_consolidado_ex12)}")
print(f"Jogadores que subiram: {(df_consolidado_ex12['tendencia_overall'] == 'subiu').sum()}")
print(f"Jogadores que caíram: {(df_consolidado_ex12['tendencia_overall'] == 'caiu').sum()}")
print(f"Jogadores estáveis: {(df_consolidado_ex12['tendencia_overall'] == 'estavel').sum()}")
print("=" * 80)
```
